```python
interior_r = 2
sectional_r = 2

maximum = 20*np.pi
theta = np.linspace(-maximum, maximum , 200)
alpha = np.linspace(0, 2 * np.pi, 200)

theta, alpha = np.meshgrid(theta, alpha)

f = interior_r + sectional_r * np.cos(alpha)
    

x = f*np.cos(theta)
y = f*np.sin(theta)
z = sectional_r * np.sin(alpha)
```

```python
figure = plt.figure(figsize=(10,10))
ax1 = plt.subplot(projection = "3d")
ax1.plot_surface(x,y,z, cmap = "viridis")

plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax4 = plt.subplot(projection = "3d")

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x, y = np.meshgrid(x,y)

z = np.sin(np.sqrt(x**2 + y**2))

surface = ax4.plot_surface(x,y,z, cmap = "plasma")
fig.colorbar(surface, shrink = 1, aspect = 5, )
plt.show()
```

```python
theta = np.linspace(-maximum, maximum, 200)
alpha = np.linspace(0, 2 * np.pi, 200)

theta, alpha = np.meshgrid(theta, alpha)

interior_r = 10
sectional_r = 3
b = 0.2

x = (interior_r + sectional_r * np.cos(alpha)) * np.cos(theta) * np.exp(b*theta)
y = (interior_r + sectional_r * np.cos(alpha)) * np.sin(theta) * np.exp(b*theta)
z = sectional_r * np.sin(alpha) * np.exp(b*theta)

figure1  = plt.figure(figsize=(10,10))

ax1 = plt.subplot(projection = "3d")
surf = ax1.plot_surface(x,y,z, cmap = "viridis")
figure1.colorbar(surf, shrink = 1, aspect = 5)

ax1.set_title("Log horn")

plt.show()
```
