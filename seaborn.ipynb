```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic_latest.csv')
sns.histplot(data["Age"], kde=True, bins=40, color="red", edgecolor = "blue")
plt.title("Age and count")
plt.show()
```

```python
sns.countplot(x ="Sex", hue="Survived", data= data)
plt.title("Sex vs Survived")
plt.show()
```

```python
gender_survival = data.groupby("Sex")["Survived"].mean().reset_index()

sns.barplot(x ="Sex", y = "Survived", data= gender_survival, palette= "Set2")
plt.title("% Survived by gender")
plt.ylim(0, 1)
plt.show()
```

```python
df = data[data["Fare"]<300]

#sns.boxplot(x = "Pclass", y = "Fare", data = df, palette = "mako")
sns.violinplot(x = "Sex", y = "Fare", data = df, palette = "mako")
plt.show()
```

```python
sns.pairplot(data[["Age", "Fare", "Survived"]], hue = "Survived")
#plt.legend(labels = ["Survived", "Did not survive"])
plt.suptitle("Pair plot", y = 1.05)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.displot(x  = "Fare", 
            hue = "Survived", 
            data =df,
            kind="kde", height=6,
            multiple="fill", clip=(0, None),
            palette="ch:rot=-.25,hue=1,light=.75")
```
